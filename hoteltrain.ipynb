```python
tags=data.Tags
allTags = []

for i in tags:
    splitted = i.split(',')
    for j in splitted :
        eachsplit=j.split("'")
        for k in eachsplit :
            if (len(k.strip())>1) :
                if k not in allTags :
                    allTags.append(k)

print(allTags)
```
